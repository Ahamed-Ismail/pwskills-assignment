## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in various formats. It provides a wide range of functions and classes for generating high-quality plots, charts, histograms, scatter plots, and more.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

1. Line Plot: A line plot is used to display data points connected by straight line segments. It is useful for showing trends and patterns over time or continuous variables.

2. Bar Plot: A bar plot (or bar chart) represents categorical data with rectangular bars, where the length or height of each bar corresponds to the value being represented. It is commonly used to compare values across different categories.

3. Scatter Plot: A scatter plot displays individual data points as dots on a 2D plane, with one variable plotted on the x-axis and another on the y-axis. It is useful for visualizing the relationship between two continuous variables.

4. Histogram: A histogram represents the distribution of a continuous variable by dividing it into intervals (bins) and displaying the frequency or proportion of data points falling into each bin. It helps understand the data's underlying distribution.

5. Pie Chart: A pie chart displays data as slices of a circular pie, where each slice represents a category and its size corresponds to the proportion or percentage it represents. It is suitable for visualizing proportions or compositions of different categories.


## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np

np.random.seed(3)

## Note: Also add title, xlabel, and ylabel to the plot.


A scatter plot is a type of plot that displays individual data points as dots on a two-dimensional plane. It is useful for visualizing the relationship between two continuous variables.

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.xlabel('x value')
plt.ylabel('y label')
plt.title('x vs y')
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
## Use the following data:
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])


x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])


x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])


x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot Line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title("Line 1")

# Plot Line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title("Line 2")

# Plot Line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title("Line 3")

# Plot Line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title("Line 4")

# Adjust the spacing between subplots
plt.tight_layout()

plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


Bar Plot: A bar plot (or bar chart) represents categorical data with rectangular bars, where the length or height of each bar corresponds to the value being represented. It is commonly used to compare values across different categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.show()

plt.barh(company,profit)
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays a summary of the data's minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum values, as well as any potential outliers.

Box plots are useful for understanding the spread, skewness, and potential outliers in a dataset. They provide a concise visualization of the data distribution and facilitate comparisons between different groups or categories.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Boxes")
plt.ylabel("Values")

plt.show()
```
