## 13. What are the benefits of using Python?

Easy to learn and read: Python has a simple and straightforward syntax that is easy to learn, read, and write. This makes it an ideal language for beginners and for rapid prototyping.

## 14. How is memory managed in Python?

Memory management in Python is handled by the Python interpreter and its underlying memory management system. Python uses automatic memory management, which means that the allocation and deallocation of memory is handled automatically by the interpreter, without the need for explicit memory management by the programmer.

The Python interpreter uses a technique called reference counting to manage memory. Every object in Python has a reference count, which is the number of references to the object in the program. When an object is no longer referenced by any part of the program, its reference count drops to zero, and the interpreter automatically frees the memory associated with the object.

## 15. How to install Python on Windows and set path variables?

Download the Python installer from the official Python website (https://www.python.org/downloads/windows/). Make sure to download the correct version for your system (32-bit or 64-bit).

Run the installer and follow the instructions to install Python on your system. Make sure to select the "Add Python to PATH" option during the installation process.

Once Python is installed, you can test it by opening a command prompt and typing "python" followed by the Enter key. This should launch the Python interpreter.

If you receive an error message indicating that "python" is not recognized as a command, you may need to add Python to your system's path variables. To do this, follow these steps:

a. Right-click on the "Start" button and select "System" from the menu.

b. Click on "Advanced system settings" in the left-hand pane.

c. Click on the "Environment Variables" button.

d. Under "System Variables", scroll down and find the "Path" variable. Click on "Edit".

e. Click on "New" and add the path to the Python installation directory (e.g., "C:\Python38"). Click "OK" to close all of the windows.

Close the command prompt and open a new one. Type "python" again to launch the Python interpreter. This time, it should launch without any errors.


## 16. Is indentation required in python?

Yes, indentation is required in Python. In fact, indentation is a fundamental part of the Python language syntax and is used to define the structure and hierarchy of code blocks.

In Python, code blocks are defined by their indentation level, rather than by using explicit keywords like "begin" or "end". This means that the indentation of each line of code within a block must be consistent.
